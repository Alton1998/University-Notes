```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import scipy
```

```python
PATH = "/content/Image_HM4D_3.jpg"
```

```python
image_file = plt.imread(PATH)
```

```python
plt.imshow(image_file)
```

## Fourier Transform

```python
ft = scipy.fft.fft2(image_file)
```

```python
plt.imshow(np.abs(ft),norm=LogNorm(vmin=5))
plt.colorbar()
```

## Centralized Fourier Transform

```python
ft = scipy.fft.fftshift(ft)
```

```python
plt.imshow(np.abs(ft),norm=LogNorm(vmin=5))
plt.colorbar()
```
