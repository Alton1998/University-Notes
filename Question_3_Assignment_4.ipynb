```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import scipy.signal
from scipy.fft import fftshift, fft2,ifftshift,ifft2
from matplotlib.colors import LogNorm
import time
```

```python
pool_image = plt.imread("pool.jpg")
```

```python
plt.title("Original Pool Image")
plt.imshow(pool_image)
```

```python
transformed = fft2(pool_image)
```

```python
fft_pool_image = fftshift(transformed)
n_row, n_col=fft_pool_image.shape
mask = np.ones((n_row,n_col))
mask[330:375,530:575] = 0
mask[430:475,630:675] = 0
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1, 2, 1)
plt.title("Centralized Fourier Spectrum")
plt.imshow(np.abs(fft_pool_image),norm=LogNorm(vmin=10**5))
plt.colorbar()
plt.subplot(1,2,2)
plt.title("Proposed Band Stop Mask")
plt.imshow(mask)
plt.colorbar()
```

```python
filtered_fft_amplitude_spectrum = np.copy(fft_pool_image)
filtered_fft_amplitude_spectrum[330:375,530:575] = 0
filtered_fft_amplitude_spectrum[430:475,630:675] = 0
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1, 2, 1)
plt.title("Original Centralized Fourier Spectrum")
plt.imshow(np.abs(fft_pool_image),norm=LogNorm(vmin=10**5))
plt.colorbar()
plt.subplot(1,2,2)
plt.title("Filtered Centralized Fourier Spectrum")
plt.imshow(np.abs(filtered_fft_amplitude_spectrum),norm=LogNorm(vmin=10**5))
plt.colorbar()
```

```python
filtered_image = ifft2(ifftshift(filtered_fft_amplitude_spectrum))
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1, 2, 1)
plt.title("Original Pool Image")
plt.imshow(pool_image)
plt.colorbar()
plt.subplot(1,2,2)
plt.title("Filtered Image")
plt.imshow(np.abs(filtered_image))
plt.colorbar()
```
