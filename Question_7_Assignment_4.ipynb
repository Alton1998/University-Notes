```python
plt.figure(figsize=(20,6))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(image)
plt.subplot(1,2,2)
plt.title("Modified Image")
plt.imshow(modified_image_matrix)
```

```python
original_image_energy = np.sum(np.square(np.abs(image)))
modified_image_energy= np.sum(np.square(np.abs(modified_image_matrix)))
print(original_image_energy)
print(modified_image_energy)
```

```python
noise_matrix = np.subtract(image,modified_image_energy)
```

```python
noise_energy = np.sum(np.square(np.abs(noise_matrix)))
```

```python
SNR_FOR_ORIGINAL_IMAGE = original_image_energy/noise_energy
SNR_FOR_MODIFIED_IMAGE = modified_image_energy/noise_energy

print(SNR_FOR_ORIGINAL_IMAGE)
print(SNR_FOR_MODIFIED_IMAGE)
```

```python
SNR_FOR_ORIGINAL_IMAGE_DB = 10*math.log10(SNR_FOR_ORIGINAL_IMAGE)
SNR_FOR_MODIFIED_IMAGE_DB = 10*math.log10(SNR_FOR_MODIFIED_IMAGE)

print(SNR_FOR_ORIGINAL_IMAGE_DB)
print(SNR_FOR_MODIFIED_IMAGE_DB)
```
