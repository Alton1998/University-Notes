```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
```

```python
image=plt.imread("GkTemple1200x800g.jpg")
```

```python
plt.imshow(image)
```

```python
discrete_wavelet_transform_coeffs = pywt.wavedec2(image,wavelet="haar",level=2)
```

```python
len(discrete_wavelet_transform_coeffs)-1
```

```python
approximation_coeffs = discrete_wavelet_transform_coeffs[0]
```

```python
a_row,a_col = approximation_coeffs.shape
```

```python
plt.title(f"Approximation Coefficients {a_row} x {a_col}")
plt.imshow(approximation_coeffs)
```

```python
detail_coeff_1 = discrete_wavelet_transform_coeffs[1]
```

```python
detail_coeff_1_horiontal = detail_coeff_1[0]
detail_coeff_1_vertical = detail_coeff_1[1]
detail_coeff_1_diagonal = detail_coeff_1[2]
```

```python
detail_coeff_0 = discrete_wavelet_transform_coeffs[2]
```

```python
detail_coeff_0_horiontal = detail_coeff_0[0]
detail_coeff_0_vertical = detail_coeff_0[1]
detail_coeff_0_diagonal = detail_coeff_0[2]
```

```python
dc1_h_row, dc1_h_col=detail_coeff_1_horiontal.shape
dc1_v_row, dc1_v_col= detail_coeff_1_vertical.shape
dc1_d_row, dc1_d_col= detail_coeff_1_diagonal.shape
```

```python
dc0_h_row, dc0_h_col=detail_coeff_0_horiontal.shape
dc0_v_row, dc0_v_col= detail_coeff_0_vertical.shape
dc0_d_row, dc0_d_col= detail_coeff_0_diagonal.shape
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1,3,1)
plt.title(f"Detail 1 Horizontal coeff {dc1_h_row} x {dc1_h_col}")
plt.imshow(detail_coeff_1_horiontal)
plt.subplot(1,3,2)
plt.title(f"Detail 1 Verical coeff {dc1_v_row} x {dc1_h_col}")
plt.imshow(detail_coeff_1_vertical)
plt.subplot(1,3,3)
plt.title(f"Detail 1 Diagonal coeff {dc1_d_row} x {dc1_d_col}")
plt.imshow(detail_coeff_1_diagonal)
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1,3,1)
plt.title(f"Detail 0 Horizontal coeff {dc0_h_row} x {dc0_h_col}")
plt.imshow(detail_coeff_0_horiontal)
plt.subplot(1,3,2)
plt.title(f"Detail 0 Verical coeff {dc0_v_row} x {dc0_h_col}")
plt.imshow(detail_coeff_0_vertical)
plt.subplot(1,3,3)
plt.title(f"Detail 0 Diagonal coeff {dc0_d_row} x {dc0_d_col}")
plt.imshow(detail_coeff_0_diagonal)
```
