```python
import numpy as np
from scipy.fft import fft2,rfft2,fftshift
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import cv2
from scipy import signal
```

# Part 1

```python
filter = [[-1,-1,-1],[0,0,0],[1,1,1]]
```

```python
filter = np.array(filter)
```

```python
test_image = plt.imread("test.png")
```

```python
test_image = cv2.cvtColor(test_image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(test_image,cmap="gray")
```

```python
fft2_image=fft2(test_image)
```

```python
fft2_image.shape
```

```python
plt.imshow(np.abs(fft2_image),norm=LogNorm(vmin=5))
plt.colorbar()
```

```python
filtered_image = signal.convolve2d(test_image, filter, boundary='symm', mode='same')
```

```python
plt.imshow(filtered_image)
```

```python
plt.imshow(np.abs(fftshift(fft2(filtered_image))))
```
